## Filter to Valid Neighborhoods

```python
crime_nbhd = crime.dropna(
    subset=["NeighborhoodInfoNAME", "NeighborhoodInfoPOPULATION"]
).copy()

crime_nbhd["NeighborhoodInfoPOPULATION"] = pd.to_numeric(
    crime_nbhd["NeighborhoodInfoPOPULATION"], errors="coerce"
)

crime_nbhd = crime_nbhd.dropna(subset=["NeighborhoodInfoPOPULATION"])

crime_nbhd.shape
```

## Create Time Features

```python
crime_nbhd["DateTime"] = pd.to_datetime(crime_nbhd["DateReport"], errors="coerce")

crime_nbhd["Date"] = crime_nbhd["DateTime"].dt.date
crime_nbhd["Hour"] = crime_nbhd["DateTime"].dt.hour
crime_nbhd["DayOfWeek"] = crime_nbhd["DateTime"].dt.day_name()

def time_of_day(h):
    if pd.isna(h):
        return "Unknown"
    h = int(h)
    if 5 <= h < 11:
        return "Morning"
    elif 11 <= h < 16:
        return "Afternoon"
    elif 16 <= h < 21:
        return "Evening"
    else:
        return "Night"

crime_nbhd["TimeOfDay"] = crime_nbhd["Hour"].apply(time_of_day)

crime_nbhd[["RecordID", "Offense", "DateReport", "Date", "Hour", "DayOfWeek", "TimeOfDay"]].head()
```

## Neighborhood-Level Aggregation 

```python
nbhd_summary = (
    crime_nbhd
    .groupby("NeighborhoodInfoNAME", as_index=False)
    .agg(
        total_incidents=("RecordID", "count"),
        population=("NeighborhoodInfoPOPULATION", "max"),
        n_beats=("NeighborhoodInfoBEAT_NO", "nunique"),
        n_districts=("NeighborhoodInfoDISTRICT", "nunique")
    )
)

nbhd_summary["incidents_per_1000"] = (
    nbhd_summary["total_incidents"] / nbhd_summary["population"] * 1000
)

nbhd_summary.head(), nbhd_summary.shape
```

## Visualizations 

```python
# Top 10 neighborhoods by incidents per 1,000 residents
top10 = nbhd_summary.sort_values("incidents_per_1000", ascending=False).head(10)

plt.figure(figsize=(9, 6))
sns.barplot(data=top10,
            x="incidents_per_1000",
            y="NeighborhoodInfoNAME")
plt.title("Top 10 Neighborhoods by Incidents per 1,000 Residents")
plt.xlabel("Incidents per 1,000 residents")
plt.ylabel("Neighborhood")
plt.tight_layout()
plt.show()
```

```python
# Incidents by hour of day
plt.figure(figsize=(9, 5))
crime_nbhd.groupby("Hour")["RecordID"].count().plot(kind="line", marker="o")
plt.title("Crime Incidents by Hour of Day")
plt.xlabel("Hour (0â€“23)")
plt.ylabel("Number of incidents")
plt.tight_layout()
plt.show()
```

```python
nbhd_summary.to_csv("Neighborhood_Summary.csv", index=False)
"Saved Neighborhood_Summary.csv"
```

```python
corr_vars = [
    "total_incidents",
    "population",
    "incidents_per_1000",
    "n_beats",
    "n_districts",
]

corr = nbhd_summary[corr_vars].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap of Neighborhood Crime Features")
plt.tight_layout()
plt.show()
```

- weak positive relationship between population and total incidents 
- low correlation between population and per-capita crime
